```python
import yfinance as yf
import seaborn as sns
from keras.models import Sequential
from sklearn.preprocessing import StandardScaler
from keras.layers import LSTM
from keras.layers import Dense ,Dropout
import pandas as pd
import numpy as np
from sklearn.metrics import r2_score,mean_absolute_error,accuracy_score, mean_squared_error
from pandas.tseries.offsets import CustomBusinessDay
import matplotlib.pyplot as plt
import datetime as dt
import tensorflow as tf
import pickle
```

```python
ticker='TATAMOTORS.NS'
```

```python
ticker_obj = yf.Ticker(ticker)
T_DF=ticker_obj.history(period='1y')
```

```python
T_DF.tail()
```

```python
T_DF.drop(['Dividends','Stock Splits'],axis=1,inplace=True)
T_DF=T_DF.reset_index()
```

```python
T_DF=T_DF[T_DF['Date']>'2000-01-01']
T_dates = T_DF['Date']
```

```python
Train_data=(T_DF[T_DF['Date']<'2022-11-07']).drop('Date',axis=1)
```

```python
scaler = StandardScaler()
scaler = scaler.fit(Train_data)
Train_data_scaled = scaler.transform(Train_data)
```

```python
trainX=[]
trainY_open=[]
trainY_high=[]
trainY_low=[]
trainY_close=[]
def Pre_multivareat():
    
    n_future=1
    n_past=14
    for i in range(n_past, len(Train_data_scaled) - n_future +1):
        trainX.append(Train_data_scaled[i - n_past:i, 0:Train_data.shape[1]])
        trainY_open.append(Train_data_scaled[i + n_future - 1:i + n_future, 0])
        trainY_high.append(Train_data_scaled[i + n_future - 1:i + n_future, 1])
        trainY_low.append(Train_data_scaled[i + n_future - 1:i + n_future, 2])
        trainY_close.append(Train_data_scaled[i + n_future - 1:i + n_future, 3])
```

```python
Pre_multivareat()
trainX,trainY_open,trainY_high,trainY_low,trainY_close= np.array(trainX),np.array(trainY_open),np.array(trainY_high),np.array(trainY_low),np.array(trainY_close)
```

```python
trainX.shape
```

```python
trainX
```

```python
model_open = Sequential()
model_open.add(LSTM(64, activation='relu', input_shape=(trainX.shape[1],trainX.shape[2]) ,return_sequences=True))
model_open.add(LSTM(32, activation= 'relu', return_sequences= False))
model_open.add(Dropout(0.2))
model_open.add(Dense(trainY_open.shape[1]))

model_open.compile(loss='mse',optimizer='adam')
```

```python
model_high = Sequential()
model_high.add(LSTM(64, activation='relu', input_shape=(trainX.shape[1],trainX.shape[2]) ,return_sequences=True))
model_high.add(LSTM(32, activation= 'relu', return_sequences= False))
model_high.add(Dropout(0.2))
model_high.add(Dense(trainY_high.shape[1]))

model_high.compile(optimizer='adam', loss='mse')
```

```python
model_low = Sequential()
model_low.add(LSTM(64, activation='relu', input_shape=(trainX.shape[1],trainX.shape[2]) ,return_sequences=True))
model_low.add(LSTM(32, activation= 'relu', return_sequences= False))
model_low.add(Dropout(0.2))
model_low.add(Dense(trainY_low.shape[1]))

model_low.compile(optimizer='adam', loss='mse')
```

```python
model_close = Sequential()
model_close.add(LSTM(64, activation='relu', input_shape=(trainX.shape[1],trainX.shape[2]) ,return_sequences=True))
model_close.add(LSTM(32, activation= 'relu', return_sequences= False))
model_close.add(Dropout(0.2))
model_close.add(Dense(trainY_close.shape[1]))

model_close.compile(optimizer='adam', loss='mse')
```

```python
holidays = ['2022-01-26','2022-03-01','2022-03-18','2022-04-14','2022-04-15','2022-05-03','2022-08-09','2022-08-15','2022-08-31','2022-10-05','2022-10-26','2022-11-08']
buz_day=CustomBusinessDay(holidays=holidays)
```

```python
today=dt.date.today()
```

```python
def model_tuning(epochs,batch_size,validation_split,verbose):
    tf.config.run_functions_eagerly(True)
    print('Opening sctock price training Started....')
    history_open = model_open.fit(trainX, trainY_open, epochs=epochs, batch_size= batch_size, validation_split= validation_split, verbose =verbose )
    print('Opening sctock price training Ended....')
    print('High sctock price training Started....')
    history_high = model_high.fit(trainX, trainY_high, epochs=epochs, batch_size= batch_size, validation_split= validation_split, verbose =verbose )
    print('High sctock price training Ended....')
    print('Low sctock price training Started....')
    history_low = model_low.fit(trainX, trainY_low, epochs=epochs, batch_size= batch_size, validation_split= validation_split, verbose =verbose )
    print('Low sctock price training Ended....')
    print('Closeing sctock price training Started....')
    history_close = model_close.fit(trainX, trainY_close, epochs=epochs, batch_size= batch_size, validation_split= validation_split, verbose =verbose )
    print('Closeing sctock price training Ended....')
    
    n_feature = 30
    forcast_dates = pd.date_range('2022-10-05', periods= n_feature, freq=buz_day).tolist()
    n_days_for_prediction=n_feature
    
    prediction_open = model_open.predict(trainX[-n_days_for_prediction:])
    prediction_high = model_high.predict(trainX[-n_days_for_prediction:])
    prediction_low = model_low.predict(trainX[-n_days_for_prediction:])
    prediction_close = model_close.predict(trainX[-n_days_for_prediction:])
    
    prediction_copies_open = np.repeat(prediction_open, Train_data.shape[1], axis=-1)
    y_pred_future_open = scaler.inverse_transform(prediction_copies_open)[:,0]
    prediction_copies_high = np.repeat(prediction_high, Train_data.shape[1], axis=-1)
    y_pred_future_high = scaler.inverse_transform(prediction_copies_high)[:,0]
    prediction_copies_low = np.repeat(prediction_low, Train_data.shape[1], axis=-1)
    y_pred_future_low = scaler.inverse_transform(prediction_copies_low)[:,0]
    prediction_copies_close = np.repeat(prediction_close, Train_data.shape[1], axis=-1)
    y_pred_future_close = scaler.inverse_transform(prediction_copies_close)[:,0]
    
    df_forcast = pd.DataFrame({'Date':np.array(forcast_dates),'Open':(y_pred_future_open),'High':(y_pred_future_high),'Low':(y_pred_future_low),'Close':(y_pred_future_close)})
    df_forcast['Date']=pd.to_datetime(df_forcast['Date'])
    global final_df
    final_df = df_forcast
    final_df['Actual_open']=T_DF['Open'].tail(n_feature).values
    final_df['Actual_high']=T_DF['High'].tail(n_feature).values
    final_df['Actual_low']=T_DF['Low'].tail(n_feature).values
    final_df['Actual_close']=T_DF['Close'].tail(n_feature).values
    
    final_df['Diff_open'] = final_df['Actual_open']-final_df['Open']
    final_df['Diff_high'] = final_df['Actual_high']-final_df['High']
    final_df['Diff_low'] = final_df['Actual_low']-final_df['Low']
    final_df['Diff_high'] = final_df['Actual_close']-final_df['Close']
    
    final_df['Diff_Percent_open']= ((final_df['Diff_open']/final_df['Actual_open'])*100)
    final_df['Diff_Percent_high']= ((final_df['Diff_high']/final_df['Actual_high'])*100)
    final_df['Diff_Percent_low']= ((final_df['Diff_low']/final_df['Actual_low'])*100)
    final_df['Diff_Percent_close']= ((final_df['Diff_high']/final_df['Actual_close'])*100)
    
    Avg_diff_open = np.average(final_df['Diff_Percent_open'])
    Avg_diff_high = np.average(final_df['Diff_Percent_high'])
    Avg_diff_low = np.average(final_df['Diff_Percent_low'])
    Avg_diff_close = np.average(final_df['Diff_Percent_close'])
    
    print(final_df)
    print(Avg_diff_open)
    print(Avg_diff_high)
    print(Avg_diff_low)
    print(Avg_diff_close)
```

```python
for i in [1]:
    epochs=48
    batch_size=12
    validation_split=0.1
    verbose=1
    model_tuning(epochs,batch_size,validation_split,verbose)
    print('Loop Completed....')
```
